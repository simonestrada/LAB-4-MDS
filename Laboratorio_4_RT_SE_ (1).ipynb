En base a lo realizado anteriormente, realice una limpieza de los dataset. Asuma los supuestos se침alados en la pregunta anterior para limpiar el `DataFrame` y cree una funci칩n que le permita limpiar los datos. Habiendo limpiado el `DataFrame`, compruebe nuevamente los datos con la funci칩n creada en la secci칩n 1.1.

**To-Do:**

- [ ] Crear funci칩n que permita limpiar el dataset.

**Hint**

Puede ser 칰til los m칠todos ```.drop_duplicates()``` y ```.dropna()``` como tambi칠n utilizar queries con m치scaras booleanas.


**Respuesta:**

````python
def clean_data(df):


  """
  Igual que en la funcion se pasa a numerico para poder obtener las columnas numericas del dataset.

  Luego se pasan estas a numericas en el df original, para luego poder filtrar las filas que estan en el rango de notas adecuado.
  Posteriormente se dropean duplicados en la columna "names" ya que no pueeden haber nombres repetidos y se dropean los nan.
  """
  numeric_df = df.apply(pd.to_numeric, errors='coerce')
  numeric_df= numeric_df.dropna(axis=1, how='all')
  cols=numeric_df.columns

  for i in cols:
    df[i]=pd.to_numeric(df[i])

  for i in cols:
    df = df[(df[i] >= 0) & (df[i] <= 100)]
  df=df.drop_duplicates(subset="names")
  df=df.dropna()

  return df
````

### 1.3.3 Reprobados y Destacados [1.5 Punto, 0.5 Cada DataFrame]

Contentos con su desempe침o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a침o y los que han tenido desempe침o de destacado. Para esto se considerar치 que un alumno ha reprobado el a침o cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerar치 destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una funci칩n que permita obtener tres Dataframes:
- El primero deber치 poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero ser치 un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el a침o.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):
  """
  primero se filtran los reprobados con GPA menor a 4
  despues se obtienen los destacados con un procedimiento parecido y luego se sortean en orden descendente.
  Finalmente para el df similar, se copia el original y utilizando reprobados como condicion se cambia los valores de GPA en estas filas a "R".
  """
  reprobados = dataframe_in[dataframe_in["GPA"] < 4.0]

  destacados = dataframe_in[(dataframe_in["math score"] > 5.8) & (dataframe_in["reading score"] > 5.8) & (dataframe_in["writing score"] > 5.8)]
  destacados = destacados.sort_values("GPA", ascending = False)
  similar = dataframe_in.copy()
  similar.loc[reprobados.index,"GPA"] = "R"

  return reprobados, similar, destacados
````

### 1.3.5 Quintiles de Rendimiento [0.5 Puntos]

El rector, intrigado por saber si existe una relaci칩n entre el nivel educacional de los padres y las notas de sus hijos les plantea estudiar dicha relaci칩n. Para esto, les solicita que desde el `DataFrame` que posee a todos los alumnos (`DataFrame` obtenido tras la limpieza), generen una nueva columna que clasifique en quintiles de rendimiento para sus alumnos: `['Baj칤simo', 'Bajo', 'Medio', 'Alto', 'Muy alto']`. (investigar `qcut()`)

Hecho esto, grafique a trav칠s de un scatter plot el nivel educacional de los padres v/s el promedio de los alumnos, utilice los cuantiles generados anteriormente para mostrar en que cuantil se encuentra cada uno de los datos.

Realizado el grafico, responda la siguiente pregunta: **쮼s posible visualizar alguna tendencia en el gr치fico?.**



**Respuesta:**

````python
def cuantiles(df):


  """
  Se utiliza la funcion qcut, para separar el df en cuantiles respecto a la columna GPA.
  """
  df["GPA cuantiles"]= pd.qcut(
    df["GPA"],
    q=5,
    labels=['Baj칤simo', 'Bajo', 'Medio', 'Alto', 'Muy alto'],
    )

  return df

cuantiles(df_grades)

fig = px.scatter(
    df_grades.sort_values(by="GPA"),
    x="GPA",
    y="parental level of education",
    hover_data=["gender"],
    color="GPA cuantiles",
    title="GPA por Nivel de educaci칩n de los padres",
    color_discrete_sequence=["tomato", "orange", "yellow", "greenyellow", "green"]
)

fig
````
